```python
import string 
def calculate_frequencies(file_contents):
    word_frequencies = {}
    #punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
    uninteresting_words = ["the", "a", "to", "if", "is", "it", "of", "and", "or", "an", "as", "i", "me", "my", \
    "we", "our", "ours", "you", "your", "yours", "he", "she", "him", "his", "her", "hers", "its", "they", "them", \
    "their", "what", "which", "who", "whom", "this", "that", "am", "are", "was", "were", "be", "been", "being", \
    "have", "has", "had", "do", "does", "did", "but", "at", "by", "with", "from", "here", "when", "where", "how", \
    "all", "any", "both", "each", "few", "more", "some", "such", "no", "nor", "too", "very", "can", "will", "just"]

    for letter in file_contents.split(" "):
        if letter.isalpha() and letter not in uninteresting_words:
            if letter not in word_frequencies:
                word_frequencies[letter] = 1
            else:
                word_frequencies[letter] += 1 

    #wordcloud
    cloud = wordcloud.WordCloud()
    cloud.generate_from_frequencies(word_frequencies)
    return cloud.to_array()
    



# data = open("amidsummernightdream.txt")
# text_data = data.read()
# filtered_text = text_data.translate(str.maketrans('','',string.punctuation))
# res = calculate_frequencies(filtered_text)
# res
    
```
